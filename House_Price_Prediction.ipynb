```python
x=df[["date","bedrooms","bathrooms","sqft_living","sqft_lot","waterfront","view","condition","sqft_above","sqft_basement","yr_built","yr_renovated"]]
y=df["price"]
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=151)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln=LinearRegression()
```

```python
numerical_df = df.select_dtypes(include=[float, int])
non_numerical_cols = [col for col in x_train.columns if x_train[col].dtype not in ['float64', 'int64']]
for col in non_numerical_cols:
  x_train = pd.get_dummies(x_train, columns=[col]) 
ln.fit(x_train, y_train)
```

```python
ln.coef_
```

```python
print(ln.intercept_)
```

```python
print(ln.coef_.shape)
print(x.columns.shape)

print(df.index.is_unique)

ln_coef_reshaped = ln.coef_[:12].reshape(12, 1)
coeff_df = pd.DataFrame(ln_coef_reshaped, x.columns, columns=['Coefficient'])

coeff_df
```

```python
missing_features = set(x_train.columns) - set(x_test.columns)

if missing_features:
    for feature in missing_features:
        x_test[feature] = 0

x_test = x_test[x_train.columns]

predictions = ln.predict(x_test)
```

```python
predictions
```

```python
sns.scatterplot(x=predictions, y=y_test)
plt.xlabel("Predictions")
plt.title("Evaluation Of My Linear Model")
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import math
from sklearn import metrics
```

```python
print("Mean Squared Error: ", metrics.mean_squared_error(y_test, predictions))
print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, predictions))
print("Root Mean Squarred Error: ", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
